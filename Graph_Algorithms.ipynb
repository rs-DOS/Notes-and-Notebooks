Graphs are usually represented by an *adjacency matrix*. This gives us an *n x n* matrix whose *(i,j)* th entry, a<sub>ij</sub> = 0 if there is not edge between the vertices and 1 if there is a edge between the vertices. This representation consumes a lot of space and is very sparse. The advantage is that it can be checked in constant time.<br>
Another representation of graphs is using lists known as adjacency lists.

```python
#Representing graphs using adjacency matrix
import numpy as np
import networkx as nx

g1 = np.array([[0, 1, 0],#Edge b/w v1 & v2
              [0,0,1],#Edge b/w v2 & v3
              [1,0,0]#Edge b/w v3 & v1    
])
h = nx.from_numpy_matrix(g1)
nx.draw(h)
```
