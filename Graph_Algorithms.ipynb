### Representing graphs
Graphs are usually represented by an *adjacency matrix*. This gives us an *n x n* matrix whose *(i,j)* th entry, a<sub>ij</sub> = 0 if there is not edge between the vertices and 1 if there is a edge between the vertices. This representation consumes a lot of space and is very sparse. The advantage is that it can be checked in constant time.<br>


```python
#Representing graphs using adjacency matrix
import numpy as np
import networkx as nx

#Diagonal is empty as all vertices are connected to themselves
g1 = np.array([[0, 1, 0],#Edge b/w v1 & v2
              [0,0,1],#Edge b/w v2 & v3
              [1,0,0]#Edge b/w v3 & v1    
])
h = nx.from_numpy_matrix(g1)
nx.draw(h)
```

Another representation of graphs is using lists known as adjacency lists. This is more of a mapping where we have a list of vertices and for every vertex is list of vertices that are connected to it.<br>

### Path finding algorithms
A common task in many graph problems is to find paths between two vertices. A path is finite sequence of edges which joins a sequence of vertices. Finding the shortest path between two vertices is used everyday in transportation applications. <br>
This can be done as a combination of book keeping and prioritisation of which nodes to visit next. Assuming that the graph is represented as a list of edges, we pick a node and loop through all the other nodes connected to it. If the node has been visited before we ignore and move on to the next node. For every new node we visit every other node connected to it. This goes on till we run out of new nodes to visit.
